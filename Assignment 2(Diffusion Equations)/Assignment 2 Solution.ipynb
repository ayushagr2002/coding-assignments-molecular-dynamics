### Question 1.1.1

Assumption: 
- We assume that the 1D line has rightmost coordinate 100 and leftmost corrdinate -100.
- The man is initially at the origin and probability of finding him at origin is 1 and 0 everywhere else.
- The probability of finding the man beyond 100 and -100 is zero at all times.
- In each timestep, he either moves right or left by 1 step.
- All the calculations are done for 100 timesteps.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
dx = 10
dt = 1
time = n

dPdT = np.empty(2*n+1)
x = np.arange(start=-n, stop=n+1, step=1)
P = np.zeros(2*n+1)
D = dx**2/2*dt
P[n] = 1
t = np.arange(0, time, dt)
timesteps = len(t)
for i in range(0, timesteps):
    dPdT[1:-1] = D*((P[2:] - 2*P[1:-1] + P[0:-2])/dx**2) 
    dPdT[0] = D*((P[1] - 2*P[0] + 0)/dx**2)
    dPdT[2*n] = D*((0 - 2*P[2*n] + P[2*n-1])/dx**2)
    P = P + dPdT*dt
    
print(P)
```

### Question 1.1.4

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
dx = 10
dt = 1
time = n

dPdT = np.empty(2*n+1)
x = np.arange(start=-n, stop=n+1, step=1)
P = np.zeros(2*n+1)
D = dx**2/2*dt
P[n] = 1
t = np.arange(0, time, dt)

meanchart = np.empty([len(t)])
meansqrchart = np.empty([len(t)])

for i in range(0, len(t)):
    dPdT[1:-1] = D*((P[2:] - 2*P[1:-1] + P[0:-2])/dx**2) 
    dPdT[0] = D*((P[1] - 2*P[0] + 0)/dx**2)
    dPdT[2*n] = D*((0 - 2*P[2*n] + P[2*n-1])/dx**2)
    P = P + dPdT*dt
    meanchart[i] = np.sum(np.multiply(P, x))
    meansqrchart[i] = np.sum(np.multiply(P, np.square(x)))

plt.plot(np.arange(0, n), meanchart, label='Mean Displacement')
plt.plot(np.arange(0, n), meansqrchart, label='Mean Square Displacement')
plt.xlabel('Steps')
plt.legend()
plt.show()
```

- The mean displacement is always 0 irrespective of the number of steps taken.
- The mean square displacement is equal to the number of steps taken(y=x).

Assumption:
- The lower left corner of metal is origin
- The parts of the metal that are in contact with the disk have a temperature equal to that of the disk.

```python
import numpy as np
import matplotlib.pyplot as plt

Tlow = 300
Thigh = 700
x = 10
y = 10

dx = 0.1
dy = 0.1
dx2 = dx**2
dy2 = dy**2

D = 4


r = 2
circlex = 5
circley = 5

u0 = Tlow * np.ones((int(x/dx), int(y/dy)))
u = u0.copy()

for i in range(int(x/dx)):
    for j in range(int(y/dy)):
        pdist = (i*dx - circlex)**2 + (j*dy - circley)**2
        if(pdist < r**2):
            u0[i,j] = Thigh
```

```python
D = 4
dt = dx2 * dy2 / (2 * D * (dx2 + dy2))
u0 = Tlow * np.ones((int(x/dx), int(y/dy)))
u = u0.copy()

for i in range(int(x/dx)):
    for j in range(int(y/dy)):
        pdist = (i*dx - circlex)**2 + (j*dy - circley)**2
        if(pdist < r**2):
            u0[i,j] = Thigh

# Recurrence relation solver
def solve_timestep(u0, u):
    u[1:-1, 1:-1] = u0[1:-1, 1:-1] + D*dt*((u0[2: , 1:-1] - 2*u0[1:-1, 1:-1] + u0[:-2, 1:-1])/dx2 + 
                                            (u0[1:-1, 2:] - 2*u0[1:-1, 1:-1] + u0[1:-1, :-2])/dy2)
    u0 = u.copy()
    return u0, u

nsteps = 101

plt.ion()
for i in range(nsteps):
    u0, u = solve_timestep(u0, u)
    if i in [0, 20, 40, 60, 80, 100]:
        im = plt.imshow(u.copy(), cmap='hot' ,interpolation='nearest', extent=(0, 10, 0, 10) , vmin=Tlow, vmax=Thigh)
        plt.colorbar(im)
        plt.title('Time : {0:1.2f} ms'.format(i*dt*1000))
        plt.xlabel('x-coordinate')
        plt.ylabel('y-coordinate')
        plt.show()
        plt.pause(1)
        plt.clf()
```
