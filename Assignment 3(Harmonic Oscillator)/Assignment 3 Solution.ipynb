## Task 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = 6
v0 = 0
dt = 0.01
k = 2
m = 1
omega_2 = k/m
x1 = x0 + v0*dt
time = 7
timesteps = int(time/dt)
p0 = v0/m
p1 = p0 - k*x0*dt
```

```python
position = np.linspace(0, time, timesteps)
position[0] = x0
position[1] = x1

for i in range(1, timesteps-1):
    position[i+1] = (2 - (omega_2*dt**2))*position[i] - position[i-1]

plt.plot(np.arange(0, time, dt), position)
plt.xlabel('time')
plt.ylabel('position')
plt.show()
```

The position of the particle is oscillatory about the mean position.

```python
momentum = np.linspace(0, time, timesteps)
momentum[0] = p0
momentum[1] = p1

for i in range(1, timesteps-1):
    momentum[i+1] = (2 - ((dt**2)*k/m))*momentum[i] - momentum[i-1]

plt.plot(np.arange(0, time, dt), momentum)
plt.xlabel('time')
plt.ylabel('momentum')
plt.show()
```

## Task 3

```python
plt.plot(position, momentum)
plt.title('Phase Space')
plt.xlabel('position')
plt.ylabel('momentum')
plt.show()
```

The phase space plot is a ellipse owing to the equation: 1/2kx^2 + p^2/2m = H, since the total energy H is constant in case of 1D oscillator, it represents the equation of an ellipse

```python
import numpy as np
import matplotlib.pyplot as plt

def pe(x):
    return (0.5*k*x**2)

def dpe(x):
    return (k*x)

alpha = 0.001
tolerance = 0.0001
x = x0 - alpha*(dpe(x0))

pos = np.array([])
potene = np.array([])
pos = np.append(pos, float(x0))
potene = np.append(potene, 0.5*k*x0**2)
while(abs(x - x0) > tolerance):
    x0 = x
    x = x0 - alpha*(dpe(x0))
    pos = np.append(pos, float(x0))
    potene = np.append(potene, 0.5*k*x0**2)

print('Minimum Potential Energy: ',potene[-1])
plt.plot(pos, potene)
plt.xlabel('position')
plt.ylabel('potential energy')
plt.show()
```
